# Adjonction

```python
import pandas as pd
from home_credit.tables import Bureau, BureauBalance
from home_credit.groupby import (
    get_bureau_loan_status,
    get_bureau_mean_loan_activity,
    get_bureau_loan_status_by_month,
    get_bureau_loan_activity_by_month,
    get_rle_bureau_loan_feature_variations
)

def get_extended_clean_bureau(
    include_status: bool = True,
    include_activity: bool = True,
    include_status_variation: bool = False,
    include_activity_variation: bool = False
) -> pd.DataFrame:
    data = Bureau.clean()
    data_ext = data.reset_index()

    # TODO commencer par un concat des agrégations de bureau_balance et un seul merge
    # Gain de performances assuré
    
    # TODO gérer le problème du multi-index en colonnes
    
    # TODO étendre le mécanisme de cache aux agrégats
    
    def adj_to_data_ext(adj_data):
        on_and_how = {"on": "SK_ID_BUREAU", "how": "outer"}
        return pd.merge(left=data_ext, **on_and_how, right=adj_data)
    
    if include_status:
        status = get_bureau_loan_status(BureauBalance.clean())
        adj_to_data_ext(status)
    
    if include_activity:
        activity = get_bureau_mean_loan_activity(BureauBalance.clean())
        adj_to_data_ext(activity)
    
    if include_status_variation:
        status_by_month = get_bureau_loan_status_by_month(BureauBalance.clean())
        status_variation = get_rle_bureau_loan_feature_variations(status_by_month, "STATUS")
        adj_to_data_ext(status_variation)
    
    if include_activity_variation:
        activity_by_month = get_bureau_loan_activity_by_month(BureauBalance.clean())
        activity_variation = get_rle_bureau_loan_feature_variations(activity_by_month, "ACTIVE")
        display(activity_variation)
        
    # data_ext = pd.merge(left=data_ext, **on_and_how, right=status)
    # data_ext = pd.merge(left=data_ext, **on_and_how, right=activity)
    # data_ext = pd.merge(left=data_ext, **on_and_how, right=status_variation)
    # data_ext = pd.merge(
    #     left=data_ext, **on_and_how,
    #     right=activity_variation[[("ACTIVE", "series_rle_reduction")]]
    # )
    return data_ext.set_index(["SK_ID_CURR", "SK_ID_BUREAU"])

data_ext = get_extended_clean_bureau()
display(data_ext)
data_ext.info()
```
