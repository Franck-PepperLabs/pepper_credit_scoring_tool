````python
from home_credit.load import get_application, get_previous_application
from pepper.utils import bold
from pepper.univar import agg_value_counts

app = get_application()

print(f"{bold('n_unique')}(`application`.`SK_ID_CURR`): {app.SK_ID_CURR.nunique():n}")

prev_app = get_previous_application()
display(prev_app)

prev_app_by_curr = prev_app[["SK_ID_PREV", "SK_ID_CURR"]].groupby("SK_ID_CURR").agg("count")
# display(prev_app_by_curr)

print("n_prev_by_curr:")
display(agg_value_counts(prev_app_by_curr.SK_ID_PREV))

# demandes passées sans demande actuelle : ensemble vide, mais il fallait vérifier !
# c'est ce doit faire ressortie ma heatmap qui reste à produire
prev_app_without_app = prev_app[~prev_app.SK_ID_CURR.isin(app.SK_ID_CURR)]
display(prev_app_without_app)
````

## Empreinte mémoire vive du jeu de données

````python
from home_credit.utils import get_tables_dict
tables_dict = get_tables_dict()
````

````python
from sys import getsizeof
import numpy as np
print(getsizeof(np.int8(1)))
print(getsizeof(np.int64(1)))
````

````python
import pandas as pd
from pepper.utils import format_iB
mudts = []
for k, v in tables_dict.items():
    mudt = pd.concat([v.memory_usage(), v.dtypes], axis=1)
    mudt.loc["Index", 1] = v.index.dtype
    mudt.reset_index(inplace=True)
    mudt.columns = ["column", "bytes", "dtype"]
    mudt.insert(0, "table", k)
    mudt["bytes/n_samples"] = mudt.bytes / v.shape[0]
    mudts.append(mudt)
mudts = pd.concat(mudts, axis=0, ignore_index=True)
display(mudts)
bytes_tot = mudts.bytes.sum()
print(f"{bytes_tot:n}")
print("total memory usage:", *format_iB(bytes_tot))
````

### Groupe métier `HOUSING_PROFILE` : `_(AVG|MEDI|MODE)`

````python
from home_credit.utils import get_variables_infos

var_infos = get_variables_infos()

def _get_pat(group_name):
    pat_map = {
        "HOUSING_PROFILE": ".*_(AVG|MEDI|MODE)"
    }
    return pat_map.get(group_name)

def get_variables_description(group_name):
    pat = _get_pat(group_name)
    # TODO : mettre dans le cache
    var_infos = get_variables_infos()
    mask = var_infos.Variable.str.match(pat)
    return var_infos[mask]
    

display(get_variables_description("HOUSING_PROFILE"))
````
