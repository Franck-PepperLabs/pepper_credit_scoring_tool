Definition:
* Monthly balance snapshots of previous credit cards that the applicant has with Home Credit.
* This table has one row for each month of history of every previous credit in Home Credit (consumer credit and cash loans) related to loans in our sample – i.e. the table has (#loans in sample * # of relative previous credit cards * # of months where we have some history observable for the previous credit card) rows.

Définition :
* *Instantanés mensuels des soldes des précédentes cartes de crédit que le demandeur a avec Home Credit.*
* *Cette table a une ligne pour chaque mois de l'historique de chaque crédit précédent de cartes de crédit à Home Credit (crédit à la consommation et prêts en espèces) relatifs aux prêts dans notre échantillon - c'est-à-dire que la table a (# prêts dans l'échantillon * # de cartes de crédit précédentes relatives * # de mois où nous avons une historique observable pour la carte de crédit précédente) de lignes.*

## Rapports et relations (feat_eng)

```python
drw_data_block = data_block[data_block.columns[:-1]].dropna().astype(int)
drw_data_block.columns = ["ATM", "TOT", "OTHER", "POS"]
drw_data_block = drw_data_block[["TOT", "ATM", "POS", "OTHER"]]
display(drw_data_block)
```

Constatons que l'assertion `drw == drw_atm + drw_pos + drw_other` est vérifiée dans $100\,\%$ des cas.

```python
drw = drw_data_block.TOT
drw_atm = drw_data_block.ATM
drw_pos = drw_data_block.POS
drw_other = drw_data_block.OTHER
drw_apo = drw_atm + drw_pos + drw_other
prop_truth = drw == drw_apo
display(prop_truth.value_counts(normalize=True))
```

Quelles sont les répartitions usuelles ?

Dans le cadre du feat_eng et notamment pour l'imputation, nous nous baserons de préférence sur la médiane que sur la moyenne. En effet, la médiane est beaucoup plus robuste d'un point de vie statistique face à l'influence des outliers sur la tendance centrale.

En d'autres termes, chaque fois qu'aucune répartition ATM/POS/Other sera inconnue, nous fixerons par défaut le triplet (ATM, POS, OTHER) à (1, 0, 0).

```python
perc_atm = (drw_atm / drw).rename("% ATM")
perc_pos = (drw_pos / drw).rename("% POS")
perc_other = (drw_other / drw).rename("% OTHER")
perc = pd.concat([perc_atm, perc_pos, perc_other], axis=1)
print("medians:")
display(perc.median(axis=0))
print("means:")
display(perc.mean(axis=0))
display(perc.describe(include="all"))
```

On constate que la part d'ATM dans le nombre de retraits mensuels est :

* $17\,\%$ de tout ou rien :
    * $11\,\%$ des clients ne font que des retraits ATM
    * $6\,\%$ des client n'en font aucun
* Une relative symétrie centrale dans les rapports (rapport x ~ rapport (1 - x))

## Relations et rapports

Une fois n'est pas coutume, compte tenu du nombre de variables, de la présence de probables redondances, nous étudions d'abord les rapports et relations, pour pourvoir simplifier avant d'effectuer l'analyse statistique de la distribution des valeurs.

### Identité de `RCV` et `RCV_TOT`

La relation est vérifiée dans $99.7\,\%$ des cas, mais il y a des exceptions.

```python
rcv = rp_data_block.RCV
rcv_tot = rp_data_block.RCV_TOT
prop_truth = rcv == rcv_tot
display(prop_truth.value_counts(normalize=True))
```

Faute de meilleure définition des variables, la nature de ces exceptions reste mystérieuse.

Dans tous les cas, on vérifie l'assertion `RCV` $\le$ `RCV_TOT`.

Le rapport médian entre la différence des deux montants et le montant total exigible est de $1\,\%$ mais la moyenne est de $8\,\%$ avec un écart type de $17\,\%$ ce qui indique la présence d'outliers super-massifs. Par exemple l'enregistrement 3385916 indique un rapport de $157\,\%$.

Meilleures hypothèses à ce stade :
1. `RCV` est la dette exigible née dans le mois en cours, et `RCV_TOT` la dette exigible cumulée sur les mois précédents (le mois en cours inclus). Le caractère d'exception n'est qu'une conséquence du caractère exceptionnel des clients en situation de défaut en général, et plus particulièrement au-delà d'un mois (exception parmi les exceptions).
2. Il a ajout de pénalités prévues au contrat ou d'intérêts capitalisés.

```python
exceptions = rp_data_block[["RCV_PPL", "RCV", "RCV_TOT"]][~prop_truth]
exceptions["DIFF"] = exceptions.RCV_TOT - exceptions.RCV
exceptions["%DIFF"] = exceptions.DIFF / exceptions.RCV_TOT
exceptions.sort_values(by="%DIFF", ascending=False, inplace=True)
display(exceptions)
display(exceptions.tail(15))
```

```python
import matplotlib.pyplot as plt
display(exceptions["%DIFF"].describe(include="all"))
_ = exceptions["%DIFF"].plot.hist(bins=90, alpha=.5)
plt.yscale("log")
plt.show()
_ = exceptions["%DIFF"].plot.hist(bins=90, alpha=.5)
plt.show()
```

### Principal (capital) vs. intérêts

`RCV` ou `RCV_TOT` - `RCV_PPL` nous donne a priori le montant des intérêts.

L'étude ci-dessous montre que cette hypothèse ne tient pas.

```python
interest = rp_data_block[["RCV_PPL", "RCV", "RCV_TOT"]].copy()
interest["INT"] = interest.RCV_TOT - interest.RCV_PPL
interest["%INT"] = interest.INT / interest.RCV_PPL
interest.sort_values(by="%INT", ascending=False, inplace=True)
display(interest)
```

Cas nuls $\Leftrightarrow$ les 4 valeurs sont toutes nulles.

```python
null_interest = interest[interest["%INT"].isna()]
null_interest_assert = null_interest[null_interest.columns[:-1]] == 0
print(null_interest_assert.all().all())
display(null_interest)
```

Cas `inf` dont le sous-ensemble des cas `RCV` $\ne$ `RCV_TOT`

Ces cas restent difficiles à interpréter sans plus d'information.

Dans tous les cas `RCV_PPL` est nul, ce qui produit `inf`.

```python
import numpy as np
inf_interest = interest[np.isinf(interest["%INT"])]
inf_interest_assert = inf_interest.RCV == inf_interest.RCV_TOT # inf_interest.RCV_PPL == 0
print(inf_interest_assert.all())
display(inf_interest)
display(inf_interest[~inf_interest_assert])
```

Cas *normaux* avec un montant de capital non nul.

Ces cas ne sont pas si normaux, puisque qu'on observe deux types d'outliers :
* Des montant de capital quasi nuls et des taux d'intérêt hors norme.
* Des montant de capital négatif.

Toutes ces observations laissent entendre qu'il y a un décalage temps entre le calcul du capital restant à rembourser et celui des intérêts. Nous observons probablement le dernier mois qui finalise le remboursement d'un crédit. `RCV` ne représenterait alors pas le principal avant, mais après opération. Dans ce cas, le montant de `RCV` représente l'échéance mensuelle, avec sa part de capital et d'intérêt, mais ne contient que du capital en ce dernier mois.

N'allons pas plus loin dans ces investigations, l'étude des autres variables va probablement nous apporter les réponses qu'il nous manque.

```python
display(interest[interest.RCV_PPL != 0])
```

### Relation entre `PYT` et `PYT_TOT`

Les deux variables sont égales dans $51\,\%$ des cas.

On observe systématiquement `PYT` $\ge$ `PYT_TOT`.

**Tps** 11 min -> passé à 6 min 33 s avec le astype object

Qu'observons-nous ?

En résumé, la séquence est une forme de ce type :

[Completed|Demand]?[Active]?[Signed]?[Refused|Approved]?[Sent proposal]?

Dans les cas d'interruption de séquence :
1. si l'état avant et après est le même, on peut donc en déduire qu'il était le même entre les deux.
2. dans le cas où il est intervenu sur une transition d'état, disons qu'il sera pas prolongement extrapolé comme prolongement de l'état précédent.

La réduction des séquences peut alors se résumer à la spécification simple mais assurée suivante :
1. compléter les NA intermédiaires avec la méthode `ffill` de `fillna` (propagation avant de la dernière valeur valide).
2. réduction verticale si plusieurs lignes pour un client :
    * Active l'emporte si au moins un Active dans la colonne
    * Sinon ... voir déjà ce que 1. donne, mais en anticipant :
        * Si Completed et autre chose, l'autre chose l'emporte

#### Fonction intégrée

Le tout dans une unique fonction intégrée : confirmation des 86 s

```python
from home_credit.load import get_credit_card_balance
from pepper.pd_utils import subindex
import numpy as np
import pandas as pd


def eval_contracts_status(x):
    if pd.isnull(x):
        return np.nan
    if isinstance(x, int):
        return None
    elif "Active" in x:
        return "Active"
    elif "Signed" in x:
        return "Signed"
    elif "Completed" in x:
        return "Completed"
    else:
        return x


def pivot_group_credit_card_balance(data):
    data.MONTHS_BALANCE = -data.MONTHS_BALANCE
    var_names = list(data.columns[3:])

    # Pivot
    pivoted = data.pivot(
        index=["SK_ID_CURR", "SK_ID_PREV"],
        columns="MONTHS_BALANCE",
        values=var_names
    )
    pivoted.sort_index(inplace=True)

    # Subindex
    sk_id_curr = pivoted.index.get_level_values(0).to_series()
    subidx = subindex(sk_id_curr)
    # TODO : update subindex to avoid this :
    subidx.drop(columns=["SK_ID_CURR"], inplace=True)

    # Split: singles vs multis
    single_prev_idx = subidx[subidx.subindex == 0]
    single_prev_mask = sk_id_curr.isin(single_prev_idx.index).to_numpy()
    single_prev_pivoted = pivoted[single_prev_mask]
    
    multi_prev_idx = subidx[subidx.subindex == 1]
    multi_prev_mask = sk_id_curr.isin(multi_prev_idx.index).to_numpy()
    multi_prev_pivoted = pivoted[multi_prev_mask]

    # Singles simple reduction
    single_prev_pivoted = single_prev_pivoted.droplevel(1, axis=0)
    
    # Multis complex reduction
    grouped_multi = multi_prev_pivoted.groupby(level=0).sum()
    grouped_multi["NAME_CONTRACT_STATUS"] = \
    grouped_multi["NAME_CONTRACT_STATUS"].applymap(eval_contracts_status)

    # Union
    synth = pd.concat([single_prev_pivoted, grouped_multi], axis=0, sort=True)
    synth.sort_index(inplace=True)

    return synth

data = get_credit_card_balance().copy()
var_names = list(data.columns[3:])
synth_data = pivot_group_credit_card_balance(data)
display(synth_data)
```

### Sauvegarde

Nous avons conçu un système performant pour ne pas devoir enregistrer toutes les étapes intermédiaires en raison d'un stress excessif des capacités système (d'une unité personnelle sans superflu de puissance).

Il est donc pertinent de sauvegarder cette version certes non relationnelle, mais compressée sans perte, de l'information de suivi de la capacité du client à honorer ses dettes.

Préalablement à la sauvegarde, les tuples contenant des couples (str, int) et localement des (np.nan, int) doivent être transformés en chaînes de caractères pour pouvoir être sauvegardés.

```python
str_casted = dense_data.NAME_CONTRACT_STATUS.PROFILE.apply(str)
dense_data.loc[:, ("NAME_CONTRACT_STATUS", "PROFILE")] = str_casted
display(dense_data.NAME_CONTRACT_STATUS.PROFILE)
```
