# Adjonction de `TARGET` à chacune des tables périphériques

Nous avons démontré dans `3_relational_schema.ipynb` que tout client référencé par un `SK_ID_CURR` dans l'une quelconque des tables périphérique est bien bien présent dans la table `application` recomposée.

Par conséquent, un `left join` sur cette clé permet d'ajouter une colonne `TARGET` à chacune de ces tables, si le besoin s'en fait sentir. Ce besoin s'est manifesté pendant l'analyse exploratoire, notamment lors de l'évaluation de la corrélation avec la cible (corrélation Tau de Kendall pondérée cf. le poids des classes) de chacune des variables étudiées (et les 346 d'entre elles l'ont été systématiquement dans `4_data_blocks.ipynb`).

La fonction `targetize` permet de construire efficacement cette la colonne `TARGET` à adjoindre.

```python
from home_credit.merge import targetize
from home_credit.load import get_previous_application
prev_app = get_previous_application().copy()
target = targetize(prev_app.SK_ID_CURR)
prev_app.insert(0, "TARGET", target)
display(prev_app)
```

```python
from home_credit.merge import targetize
from home_credit.load import get_table
from home_credit.utils import var_catdist_report

def adj_target_and_report(table_name):
    data = get_table(table_name).copy()
    target = targetize(data.SK_ID_CURR)
    data.insert(0, "TARGET", target)
    # var_catdist_report(data.TARGET)
    var_catdist_report(data.TARGET[data.TARGET > -1])
    return data

_ = adj_target_and_report("bureau")
```

## `POS_CASH_BALANCE`

### Données

```python
from home_credit.load import get_table
from pepper.utils import display_key_val
data = get_table("pos_cash_balance").copy()
data.MONTHS_BALANCE = -data.MONTHS_BALANCE
# Insert the aggregation counter
data.insert(0, "n_PREV", 1)
display_key_val("number of samples", data.shape[0])
display(data)
```

### Séparation *singles* vs *multis*

Séparation des 2-index `(SK_ID_PREV, SK_ID_CURR)` pour les *singles* (un `SK_ID_PREV` pour un `SK_ID_CURR`) vs *multis* (plusieurs `SK_ID_PREV` pour un `SK_ID_CURR`).

La séparation est justifiée par l'enjeu de performance.

Pour $103\,736$ *singles*, le `groupby` ne se justifie pas.

La réduction ne concerne que $832\,589$ *multis*, ce qui demeure un nombre conséquent ($88.9\,\%$ de l'effectif du 2-index).

#### Unicité des clés

Nous vérifions qu'il ne peut y avoir plusieurs `SK_ID_CURR` pour un `SK_ID_PREV`.

La problématique n'est donc multi indexée qu'en apparence : la clé `SK_ID_CURR` suffit pour séparer.

Nombre de `SK_ID_PREV` pour une `SK_ID_CURR` et inversement :

```python
from home_credit.merge import get_singl_and_multi_index, curr_prev_unicity_report
indexes = get_singl_and_multi_index("pos_cash_balance", "SK_ID_PREV", "SK_ID_CURR")
singl_prev_idx, multi_prev_idx, _, _ = indexes
curr_prev_unicity_report(*indexes)
```

#### Séparation de la table principale

Nous pouvons effectuer le `groupby` sur un sous-ensemble de $8\,854\,633$ enregistrements contre $10\,001\,358$ si nous n'avions pas effectué cette distinction.

```python
is_multi_mask = data.SK_ID_CURR.isin(multi_prev_idx.SK_ID_CURR)
data_multis = data[is_multi_mask]
data_singles = data[~is_multi_mask]
display(data_multis)
```

### Agrégation des *multis* : 11 s

Nous effectuons une synthèse avec le couple `(SK_ID_CURR, MONTHS_BALANCE)` comme pivot.

L'agrégation est la somme pour toutes les colonnes avec un post-traitement spécifique pour la colonne `NAME_CONTRACT_STATUS`.

`SK_ID_PREV` n'étant pas agrégée, on la retire avant opération.

Les $8\,854\,633$ enregistrements sont réduits à $7\,446\,182$.

```python
grouped_multis = (
    data_multis.drop(columns="SK_ID_PREV")
    .groupby(by=["SK_ID_CURR", "MONTHS_BALANCE"])
    .sum()
)
display(grouped_multis)
```

#### Post traitement de `NAME_CONTRACT_STATUS`

Ce post-traitement a pour but de réduire les combinaisons issues de la somme interprétée en tant concaténation sur cette colonne d'état du contrat. Le choix de la somme est délibéré : sa performance dépasse de loin toute fonction d'agrégation que l'on aurait pu définir, et le post traitement est donc le plus adapté.

```python
from home_credit.feat_eng import eval_contracts_status
grouped_multis.NAME_CONTRACT_STATUS = (
    grouped_multis.NAME_CONTRACT_STATUS
    .apply(eval_contracts_status)
)
```

### Recombinaison des deux parties

#### Alignement

Côté *singles*, l'équivalent du `groupby` consiste à supprimer la colonne `SK_ID_PREV` et à passer la colonne `SK_ID_CURR` en index.

Côté *multis*, il faut sortir la colonne `MONTHS_BALANCE` de l'index.

```python
grouped_singles = data_singles.drop(columns="SK_ID_PREV")
# Trick pour repositionner MONTHS_BALANCE en bonne position (première colonne)
grouped_singles.set_index(["MONTHS_BALANCE", "SK_ID_CURR"], drop=True, inplace=True)
grouped_singles.reset_index(level=0, inplace=True)
display(grouped_singles)
```

```python
grouped_multis.reset_index(level=1, inplace=True)
display(grouped_multis)
```

#### Empilement vertical

Il ne reste qu'à effectuer un empilement vertical à l'aide de `pd.concat` et à trier l'index (objectif de confort et de lisibilité, ce n'est pas indispensable) :

```python
import pandas as pd
grouped_data = pd.concat([grouped_multis, grouped_singles])
grouped_data.sort_index(inplace=True)
display(grouped_data)
```

### Version intégrée : 21 s

```python
from home_credit.merge import groupby_curr_months

ref_grouped_data = grouped_data
grouped_data = groupby_curr_months("pos_cash_balance")

display(grouped_data)
display((
    (grouped_data == ref_grouped_data)
    # Avoid not(NaN == NaN)
    | grouped_data.isnull() & ref_grouped_data.isnull()
).all())
```

### Agrégation RLE des variations mensuelles

```python
from home_credit.merge import groupby_curr_months
grouped_data = groupby_curr_months("pos_cash_balance")
```

Tri des données par `SK_ID_CURR`, `MONTHS_BALANCE` :

```python
sorted_data = grouped_data.copy()
sorted_data.reset_index(inplace=True)
sorted_data.sort_values(by=["SK_ID_CURR", "MONTHS_BALANCE"], inplace=True)
sorted_data.set_index("SK_ID_CURR", inplace=True)
display(sorted_data)
```

Formation des groupes :

```python
gpby = sorted_data.groupby(by="SK_ID_CURR")
```

**Tps** 7 min 40 s

```python
from pepper.feat_eng import data_rle_reduction
zz = gpby.apply(data_rle_reduction)
zz.columns = list(sorted_data.columns)
display(zz)
```

Nous vérifions qu'il ne peut y avoir plusieurs `SK_ID_CURR` pour un `SK_ID_PREV`.

#### Agrégation : 17 s

Nous effectuons une synthèse avec le couple `(SK_ID_CURR, MONTHS_BALANCE)` comme pivot.

Agrégations :
* `NUM_INSTALMENT_VERSION` et `NUM_INSTALMENT_NUMBER` : il ne sont ici qu'informatifs, et intégralement restitués dans l'autre table pivotée : le maximum sera suffisant (la dernière version, la dernier numéro d'échéance)
* `DAYS_INSTALMENT` et `DAYS_ENTRY_PAYMENT` : on peut imaginer le dernier jour, ou le jour médian, sachant que la granularité importante dans notre pivot est le mois. Nous choisissons la médiane.
* `AMT_INSTALMENT`, `AMT_PAYMENT`, `n_PREV` : la somme.

Les $13\,605\,401$ enregistrements sont réduits à $9\,477\,481$ soit un taux de compression de l'ordre de $30\,\%$.

```python
grouped_data = (
    data.drop(columns="SK_ID_PREV")
    .groupby(by=["SK_ID_CURR", "MONTHS_BALANCE"])
    .agg({
        "n_PREV": "sum",
        "NUM_INSTALMENT_VERSION": "max",
        "NUM_INSTALMENT_NUMBER": "max",
        "DAYS_INSTALMENT": "median",
        "DAYS_ENTRY_PAYMENT": "median",
        "AMT_INSTALMENT": "sum",
        "AMT_PAYMENT": "sum"
    })
)
grouped_data.reset_index(level=1, inplace=True)
display(grouped_data)
```

# Big merge

Il s'agit de joindre la table `application` avec l'ensemble de ses tables périphériques.

La variable `curr` représente la table `application` brute et `curr_` sa version étendue par l'adjonction de tous ses périphériques.

On note `|` l'opération de jonction. Dans la plupart des cas c'est une adjonction gauche, c'est-à-dire une juxtposition horizontale de tables, de la gauche vers la droite. Mais en l'espèce, ayant détecté des situations de clés présentes dans des tables étrangères périphériques qui sont absentes des clés primaires de la table principale, nous allons devoir effectuer, dans ces cas, des *outer joins*.

On note ':' l'opération d'agrégation en classes d'équivalence suite au quotient par la clé primaire d'une table quotient (par exemple `bur_bal:bur` est l'agrégation avec regroupement des lignes de `bureau` par même `SK_ID_BUR`, et `bur_bal:month:bur`) :

L'agrégation sans perte d'information naïve utilise l'agrégateur `list` ou `tuple` et sa version avancée utilise un codage RLE, bien adapté au caractère longitudinal des données agrégées (variations mensuelles de variables temporelles).

... cas de :status

La formation de `curr_` peut être décrire de la manière suivante :
* `curr_ = curr | bur_:curr | prev_:curr`
* `bur_ = bur | bur_bal:month:bur | bur_bal:status:bur`
* `prev_ = prev | pos_bal:prev | cdt_bal:prev | ins_pyt:prev`
* `pos_bal:prev = pos_bal:month:prev`
* `cdt_bal:prev = cdt_bal:month:prev`
* `ins_pyt:prev = ins_pyt:month:prev | ins_pyt:num:prev`

Cela fait donc 27 opérations élémentaires (17 n-aires) `:` et `|`.
