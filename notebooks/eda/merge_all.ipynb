# Adjonction de `TARGET` Ã  chacune des tables pÃ©riphÃ©riques

Nous avons dÃ©montrÃ© dans `3_relational_schema.ipynb` que tout client rÃ©fÃ©rencÃ© par un `SK_ID_CURR` dans l'une quelconque des tables pÃ©riphÃ©rique est bien bien prÃ©sent dans la table `application` recomposÃ©e.

Par consÃ©quent, un `left join` sur cette clÃ© permet d'ajouter une colonne `TARGET` Ã  chacune de ces tables, si le besoin s'en fait sentir. Ce besoin s'est manifestÃ© pendant l'analyse exploratoire, notamment lors de l'Ã©valuation de la corrÃ©lation avec la cible (corrÃ©lation Tau de Kendall pondÃ©rÃ©e cf. le poids des classes) de chacune des variables Ã©tudiÃ©es (et les 346 d'entre elles l'ont Ã©tÃ© systÃ©matiquement dans `4_data_blocks.ipynb`).

La fonction `targetize` permet de construire efficacement cette la colonne `TARGET` Ã  adjoindre.

```python
from home_credit.merge import targetize
from home_credit.load import get_previous_application
prev_app = get_previous_application().copy()
target = targetize(prev_app.SK_ID_CURR)
prev_app.insert(0, "TARGET", target)
display(prev_app)
```

```python
from home_credit.merge import targetize
from home_credit.load import get_table
from home_credit.utils import var_catdist_report

def adj_target_and_report(table_name):
    data = get_table(table_name).copy()
    target = targetize(data.SK_ID_CURR)
    data.insert(0, "TARGET", target)
    # var_catdist_report(data.TARGET)
    var_catdist_report(data.TARGET[data.TARGET > -1])
    return data

_ = adj_target_and_report("bureau")
```

## `POS_CASH_BALANCE`

### DonnÃ©es

```python
from home_credit.load import get_table
from pepper.utils import display_key_val
data = get_table("pos_cash_balance").copy()
data.MONTHS_BALANCE = -data.MONTHS_BALANCE
# Insert the aggregation counter
data.insert(0, "n_PREV", 1)
display_key_val("number of samples", data.shape[0])
display(data)
```

### SÃ©paration *singles* vs *multis*

SÃ©paration des 2-index `(SK_ID_PREV, SK_ID_CURR)` pour les *singles* (un `SK_ID_PREV` pour un `SK_ID_CURR`) vs *multis* (plusieurs `SK_ID_PREV` pour un `SK_ID_CURR`).

La sÃ©paration est justifiÃ©e par l'enjeu de performance.

Pour $103\,736$ *singles*, le `groupby` ne se justifie pas.

La rÃ©duction ne concerne que $832\,589$ *multis*, ce qui demeure un nombre consÃ©quent ($88.9\,\%$ de l'effectif du 2-index).

#### UnicitÃ© des clÃ©s

Nous vÃ©rifions qu'il ne peut y avoir plusieurs `SK_ID_CURR` pour un `SK_ID_PREV`.

La problÃ©matique n'est donc multi indexÃ©e qu'en apparence : la clÃ© `SK_ID_CURR` suffit pour sÃ©parer.

Nombre de `SK_ID_PREV` pour une `SK_ID_CURR` et inversement :

```python
from home_credit.merge import get_singl_and_multi_index, curr_prev_unicity_report
indexes = get_singl_and_multi_index("pos_cash_balance", "SK_ID_PREV", "SK_ID_CURR")
singl_prev_idx, multi_prev_idx, _, _ = indexes
curr_prev_unicity_report(*indexes)
```

#### SÃ©paration de la table principale

Nous pouvons effectuer le `groupby` sur un sous-ensemble de $8\,854\,633$ enregistrements contre $10\,001\,358$ si nous n'avions pas effectuÃ© cette distinction.

```python
is_multi_mask = data.SK_ID_CURR.isin(multi_prev_idx.SK_ID_CURR)
data_multis = data[is_multi_mask]
data_singles = data[~is_multi_mask]
display(data_multis)
```

### AgrÃ©gation des *multis* : 11 s

Nous effectuons une synthÃ¨se avec le couple `(SK_ID_CURR, MONTHS_BALANCE)` comme pivot.

L'agrÃ©gation est la somme pour toutes les colonnes avec un post-traitement spÃ©cifique pour la colonne `NAME_CONTRACT_STATUS`.

`SK_ID_PREV` n'Ã©tant pas agrÃ©gÃ©e, on la retire avant opÃ©ration.

Les $8\,854\,633$ enregistrements sont rÃ©duits Ã  $7\,446\,182$.

```python
grouped_multis = (
    data_multis.drop(columns="SK_ID_PREV")
    .groupby(by=["SK_ID_CURR", "MONTHS_BALANCE"])
    .sum()
)
display(grouped_multis)
```

#### Post traitement de `NAME_CONTRACT_STATUS`

Ce post-traitement a pour but de rÃ©duire les combinaisons issues de la somme interprÃ©tÃ©e en tant concatÃ©nation sur cette colonne d'Ã©tat du contrat. Le choix de la somme est dÃ©libÃ©rÃ© : sa performance dÃ©passe de loin toute fonction d'agrÃ©gation que l'on aurait pu dÃ©finir, et le post traitement est donc le plus adaptÃ©.

```python
from home_credit.feat_eng import eval_contracts_status
grouped_multis.NAME_CONTRACT_STATUS = (
    grouped_multis.NAME_CONTRACT_STATUS
    .apply(eval_contracts_status)
)
```

### Recombinaison des deux parties

#### Alignement

CÃ´tÃ© *singles*, l'Ã©quivalent du `groupby` consiste Ã  supprimer la colonne `SK_ID_PREV` et Ã  passer la colonne `SK_ID_CURR` en index.

CÃ´tÃ© *multis*, il faut sortir la colonne `MONTHS_BALANCE` de l'index.

```python
grouped_singles = data_singles.drop(columns="SK_ID_PREV")
# Trick pour repositionner MONTHS_BALANCE en bonne position (premiÃ¨re colonne)
grouped_singles.set_index(["MONTHS_BALANCE", "SK_ID_CURR"], drop=True, inplace=True)
grouped_singles.reset_index(level=0, inplace=True)
display(grouped_singles)
```

```python
grouped_multis.reset_index(level=1, inplace=True)
display(grouped_multis)
```

#### Empilement vertical

Il ne reste qu'Ã  effectuer un empilement vertical Ã  l'aide de `pd.concat` et Ã  trier l'index (objectif de confort et de lisibilitÃ©, ce n'est pas indispensable) :

```python
import pandas as pd
grouped_data = pd.concat([grouped_multis, grouped_singles])
grouped_data.sort_index(inplace=True)
display(grouped_data)
```

### Version intÃ©grÃ©e : 21 s

```python
from home_credit.merge import groupby_curr_months

ref_grouped_data = grouped_data
grouped_data = groupby_curr_months("pos_cash_balance")

display(grouped_data)
display((
    (grouped_data == ref_grouped_data)
    # Avoid not(NaN == NaN)
    | grouped_data.isnull() & ref_grouped_data.isnull()
).all())
```

### AgrÃ©gation RLE des variations mensuelles

```python
from home_credit.merge import groupby_curr_months
grouped_data = groupby_curr_months("pos_cash_balance")
```

Tri des donnÃ©es par `SK_ID_CURR`, `MONTHS_BALANCE` :

```python
sorted_data = grouped_data.copy()
sorted_data.reset_index(inplace=True)
sorted_data.sort_values(by=["SK_ID_CURR", "MONTHS_BALANCE"], inplace=True)
sorted_data.set_index("SK_ID_CURR", inplace=True)
display(sorted_data)
```

Formation des groupes :

```python
gpby = sorted_data.groupby(by="SK_ID_CURR")
```

## `INSTALLMENTS_PAYMENTS` : synthÃ¨se mensuelle

### DonnÃ©es

```python
from home_credit.load import get_table
from pepper.utils import display_key_val
gregorian_month = 365.2425 / 12
data = get_table("installments_payments").copy()
# Insert the aggregation counter
data.insert(0, "n_PREV", 1)
# Calculates MONTHS_BALANCE
data.insert(0, "MONTHS_BALANCE", -(data.DAYS_INSTALMENT // gregorian_month).astype(int))
display_key_val("number of samples", data.shape[0])
display(data)
```

### AgrÃ©gation par demande actuelle (`SK_ID_CURR`)

#### SÃ©paration *uniques* vs *multis*

Il n'y a pas lieu ici de l'effectuer, d'une part parce que les *singles* ne reprÃ©sentent que $10\,\%$ de l'affaire, mais surtout parce que ce serait une erreur logique : mÃªme avec un `SK_ID_CURR` associÃ© Ã  un `SK_ID_PREV` unique, l'analyse exploratoire a montrÃ© qu'il pouvait y avoir plusieurs Ã©chÃ©ances le mÃªme mois.

Par consÃ©quent, pour cette table, l'agrÃ©gation doit porter sur la totalitÃ© des enregistrements.

#### UnicitÃ© des clÃ©s

Nous vÃ©rifions qu'il ne peut y avoir plusieurs `SK_ID_CURR` pour un `SK_ID_PREV`.

La problÃ©matique n'est donc multi indexÃ©e qu'en apparence : la clÃ© `SK_ID_CURR` suffit pour sÃ©parer.

Nombre de `SK_ID_PREV` pour une `SK_ID_CURR` et inversement :

```python
from home_credit.merge import get_singl_and_multi_index, curr_prev_unicity_report
indexes = get_singl_and_multi_index("installments_payments", "SK_ID_PREV", "SK_ID_CURR")
singl_prev_idx, multi_prev_idx, _, _ = indexes
curr_prev_unicity_report(*indexes)
```

#### AgrÃ©gation : 17 s

Nous effectuons une synthÃ¨se avec le couple `(SK_ID_CURR, MONTHS_BALANCE)` comme pivot.

AgrÃ©gations :
* `NUM_INSTALMENT_VERSION` et `NUM_INSTALMENT_NUMBER` : il ne sont ici qu'informatifs, et intÃ©gralement restituÃ©s dans l'autre table pivotÃ©e : le maximum sera suffisant (la derniÃ¨re version, la dernier numÃ©ro d'Ã©chÃ©ance)
* `DAYS_INSTALMENT` et `DAYS_ENTRY_PAYMENT` : on peut imaginer le dernier jour, ou le jour mÃ©dian, sachant que la granularitÃ© importante dans notre pivot est le mois. Nous choisissons la mÃ©diane.
* `AMT_INSTALMENT`, `AMT_PAYMENT`, `n_PREV` : la somme.

Les $13\,605\,401$ enregistrements sont rÃ©duits Ã  $9\,477\,481$ soit un taux de compression de l'ordre de $30\,\%$.

```python
grouped_data = (
    data.drop(columns="SK_ID_PREV")
    .groupby(by=["SK_ID_CURR", "MONTHS_BALANCE"])
    .agg({
        "n_PREV": "sum",
        "NUM_INSTALMENT_VERSION": "max",
        "NUM_INSTALMENT_NUMBER": "max",
        "DAYS_INSTALMENT": "median",
        "DAYS_ENTRY_PAYMENT": "median",
        "AMT_INSTALMENT": "sum",
        "AMT_PAYMENT": "sum"
    })
)
grouped_data.reset_index(level=1, inplace=True)
display(grouped_data)
```

# Big merge

Il s'agit de joindre la table `application` avec l'ensemble de ses tables pÃ©riphÃ©riques.

La variable `curr` reprÃ©sente la table `application` brute et `curr_` sa version Ã©tendue par l'adjonction de tous ses pÃ©riphÃ©riques.

On note `|` l'opÃ©ration de jonction. Dans la plupart des cas c'est une adjonction gauche, c'est-Ã -dire une juxtposition horizontale de tables, de la gauche vers la droite. Mais en l'espÃ¨ce, ayant dÃ©tectÃ© des situations de clÃ©s prÃ©sentes dans des tables Ã©trangÃ¨res pÃ©riphÃ©riques qui sont absentes des clÃ©s primaires de la table principale, nous allons devoir effectuer, dans ces cas, des *outer joins*.

On note ':' l'opÃ©ration d'agrÃ©gation en classes d'Ã©quivalence suite au quotient par la clÃ© primaire d'une table quotient (par exemple `bur_bal:bur` est l'agrÃ©gation avec regroupement des lignes de `bureau` par mÃªme `SK_ID_BUR`, et `bur_bal:month:bur`) :

L'agrÃ©gation sans perte d'information naÃ¯ve utilise l'agrÃ©gateur `list` ou `tuple` et sa version avancÃ©e utilise un codage RLE, bien adaptÃ© au caractÃ¨re longitudinal des donnÃ©es agrÃ©gÃ©es (variations mensuelles de variables temporelles).

... cas de :status

La formation de `curr_` peut Ãªtre dÃ©crire de la maniÃ¨re suivante :
* `curr_ = curr | bur_:curr | prev_:curr`
* `bur_ = bur | bur_bal:month:bur | bur_bal:status:bur`
* `prev_ = prev | pos_bal:prev | cdt_bal:prev | ins_pyt:prev`
* `pos_bal:prev = pos_bal:month:prev`
* `cdt_bal:prev = cdt_bal:month:prev`
* `ins_pyt:prev = ins_pyt:month:prev | ins_pyt:num:prev`

Cela fait donc 27 opÃ©rations Ã©lÃ©mentaires (17 n-aires) `:` et `|`.
