# **`BUREAU`**

Caractéristiques :
* $1\,716\,428$ enregistrements.
* $17$ variables, dont $2$ **`SK`**, $4$ **`DAYS`**, $1$ **`CNT`**, $6$ **`AMT`**, $4$ **`CREDIT`**

Definition:
* All client's previous credits provided by other financial institutions that were reported to Credit Bureau (for clients who have a loan in our sample).
* For every loan in our sample, there are as many rows as number of credits the client had in Credit Bureau before the application date.

Définition :
* *Tous les précédents crédits des clients fournis par d'autres institutions financières qui ont été signalés au bureau de crédit (pour les clients qui ont un prêt dans notre échantillon).*
* *Pour chaque prêt dans notre échantillon, il y a autant de lignes que le nombre de crédits que le client avait dans le bureau de crédit avant la date de la demande.*

# `AMT_*`

Par ordre décroissant de remplissage :
* `AMT_CREDIT_SUM_OVERDUE` (aucun NA) - retard de paiement actuel
* `AMT_CREDIT_SUM` ($\varepsilon$ NA) - le capital emprunté = montant du prêt
* `AMT_CREDIT_SUM_DEBT` ($15\,\%$ de NA) - le capital restant dû
* `AMT_CREDIT_SUM_LIMIT` - ($35\,\%$ de NA) - plafond de crédit sur la carte de crédit - ne concerne que les clients qui en ont une (au moins à la connaissance du Central Bureau).
* `AMT_CREDIT_MAX_OVERDUE` - ($66\,\%$ de NA) le plus grand retard de paiement enregistré sur le prêt.. Le nombre important de NA correspond probablement à l'absence de retards, la plupart des clients n'en ayant pas.
* `AMT_ANNUITY` - ($72\,\%$ de NA) annuité : étrange d'avoir autant de NA sur une telle variable qui est un paramètre de prêt et non un indicateur de suivi. Le montant du prêt et sa durée étant connus, on peut logiquement déduire cette variable : à intégrer dans les `asserts`.

Variables dérivées envisagées:
* binarisations :
    * `AMT_CREDIT_SUM_OVERDUE` et `AMT_CREDIT_MAX_OVERDUE` : retard (> 0) ou pas (0)
    * `AMT_CREDIT_SUM_LIMIT` possède ou non une carte de crédit
* discrétisations :
    * Pour toutes les varibables faire des quantiles (grandes disparités d'échelle + influence des outliers)
* rapports :
    * restant dû / prêt
    * retard et plus grand retard / annuité
