* *Historique de remboursement pour les crédits précédemment accordés par Home Credit liés aux prêts dans notre échantillon.*
* *Il y a a) une ligne pour chaque paiement effectué et b) une ligne pour chaque paiement manqué.*
* *Une ligne correspond à un paiement d'une échéance ou à une échéance correspondant à un paiement pour un crédit Home Credit précédent lié aux prêts dans notre échantillon.*

# Agrégation

Contrairement aux autres tables secondaires de troisième niveau (en étoile en partant d'application en tant que racine de l'arbre relationnel) nommées `_balance` et qui sont un relevé d'événements sur les 8 dernières années avant la demande (96 mois), la table `installments_payments` est un état de la qualité de règlement de chacune des échéances de chacun des prêts précédents de chacun des clients. C'est donc l'information la plus atomique à propos de la fiabilité du client.

On retrouve ici le même modèle en longue traine que sur la plupart des variables du jeu de données : si la majorité des clients n'ont jamais eu que quelques échéances de quelques prêts (la médiane est à 27 échéances), certains clients ont eu plus de 350 échéances, avec un maximum à 371.

Mais si on observe les dates, on peut voir qu'il s'agit de la même fenêtre de 8 ans.

La synthèse devrait donc être du même type que celle des balances, avec un pivotement et une agrégation, mais les mois n'étant pas clairement indiqués, il faut les identifier nous mêmes avant de pouvoir pivoter.

Proto à poursuivre plus tard :
1. contreperf : on calcul 4 fois le même histo => le faire une fois (numpy.histogram) et produire 4 vues (=>bar)
2. disposant des fréquences pour chaque bin, en utilisant les points (min, freq_min), (max, freq_max) :
    * calculer les droites lin-lin, lin-log, log-lin, log-log qui ne sont chacune droite que dans leur propre quadrant
    * afficher ces 4 droites dans les 4 quadrants, ce qui sera rendu comme 1 droite et 3 courbes

```python
from typing import *
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
from pepper.utils import save_and_show

def lin_log_tetra_histplot_2(
    s: pd.Series,
    color: str = "gold",
    single_figsize: Tuple[int, int] = (4, 3),
    title: str = None,
    max_bins: int = int(8*365.25),
    draw_segment: bool = True
) -> None:
    """Plots a four histograms of the same data with two axes in linear scale
    and two axes in log scale.

    Parameters
    ----------
    s : pd.Series
        A series to plot.
    color : str, optional
        The color of the histograms. Defaults to "gold".
    single_figsize : Tuple[int, int], optional
        The size of a single subplot figure. Defaults to (4, 3).
    title : str, optional
        The title of the figure. Defaults to None.
    max_bins : in, optional
        Max number of bins
    draw_segment : bool, optional
        Whether to draw a line segment from the top of the first bin to the top
        of the last bin. Defaults to False.

    Returns
    -------
    None

    """
    # Determine the number of bins
    bins = min(int(1 + s.max() - s.min()), max_bins)

    # Create the figure and subplots
    nrows = ncols = 2
    w, h = single_figsize
    fig, axes = plt.subplots(
        nrows=nrows, ncols=ncols,
        figsize=(w * ncols, h * nrows)
    )

    # Plot the histograms
    [s.plot.hist(bins=bins, color=color, ax=ax) for ax in np.ravel(axes)]

    for ax in np.ravel(axes):
        freqs = np.array([patch.get_height() for patch in ax.patches])
        display(freqs)

    # Add gridlines to each subplot
    [ax.grid(linewidth=0.2) for ax in np.ravel(axes)]

    # Set the x- and y-axis scales for the appropriate subplots
    [ax.set_xscale("log") for ax in axes[:, 1]]
    [ax.set_yscale("log") for ax in axes[1, :]]

    # Format the y-axis tick labels for the top row of subplots
    [ax.ticklabel_format(
        axis="y", style="sci", useMathText=True, scilimits=(-2, 2)
    ) for ax in axes[0, :]]

    # Add titles to each subplot
    nm = lambda k: "log" if k else "lin"
    [[axes[i, j].set_title(f"{nm(j)}-{nm(i)}") for i in [0, 1]] for j in [0, 1]]

    # Add a title to the figure
    if title is None:
        title = "Jane DOE series" if s.name is None else s.name
    plt.suptitle(title, fontsize=15, weight="bold")

    # Adjust the spacing between the subplots and save/show the figure
    fig.tight_layout()
    save_and_show(f"linlog_tetra_hist_{title.lower()}", sub_dir="dataxplor")
```

Concernant l'intégration, nous allons donc prendre la stratégie consistant à enregistrer les informations pivotée par mois, en mappant `DAYS_INSTALMENT` sur une variable `MONTHS_BALANCE` dérivée obtenue par division entière par 30.

Nous procéderons ensuite à un pivotement et une synthèse cf. ce qui a été fait dans `credit_card_balance`.

Quelles sont ici les informations pertinentes ?

Au niveau des `DAYS`, il s'agit moins de la distance absolue à aujourd'hui que le retard pris sur l'échéance.

La première variable dérivée intéressante est donc la différence entre ces deux distances en temps à la date de référence qui est celle de la nouvelle demande de prêt. Ainsi, que ce soit hier ou il y a un an, l'important est de savoir si l'échéance a été réglée à temps, ou avec 15 jours de retard.

Ensuite, on peut envisager une valeur amortie pour ce qui est ancien.

Une approche usuelle pour effectuer ce type d'amortissement sur les séries temporelles financières est la *moyenne mobile exponentielle* : `rolling.ewm(alpha).mean`, avec alpha comme paramètre d'amortissement plus ou moins prononcé du passé.
