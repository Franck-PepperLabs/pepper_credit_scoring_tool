### Test de consistence

Les supports (`tracking_period`) et les trames de données, doivent avoir un support étendu de même taille.

```python
from home_credit.tables import BureauBalance
import pandas as pd

support = BureauBalance.rle_loan_tracking_period()
frame = BureauBalance.rle_loan_status_variation()

rle = pd.concat([support, frame], axis=1)
display(rle)
```

```python
from pepper.rle import rle_support_size

rle.columns = ["SUPPORT", "FRAME"]
rle["|S|"] = rle.SUPPORT.apply(rle_support_size)
rle["|F|"] = rle.FRAME.apply(rle_support_size)
rle["|S| - |F|"] = rle["|S|"] - rle["|F|"]
print(f"# (rle['|S| - |F|'] != 0): {(rle['|S| - |F|'] != 0).sum()}")
display(rle)
```
